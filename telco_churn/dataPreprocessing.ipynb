```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
telco_data = pd.read_csv('./Telco-Customer-Churn.partial.csv')
```

```python
telco_data.head()
```

## Churn

##### Come si può osservare in seguito i dati sono abbastanza sbilanciati perchè c'è una gande maggioranza di clienti che non se va

```python
ax = sns.countplot(x = telco_data.Churn)
plt.title('Distribution of  Churn')
plt.ylabel('N. Customers')
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/len(telco_data))
    x = p.get_x() + p.get_width() / 2.8
    y = p.get_y() + p.get_height() / 2
    ax.annotate(str(percentage), (x, y), color = 'white', fontsize = 12, weight = 'bold')
plt.show()
```

## Gender

##### Il set di dati contiene circa lo stesso numero di clienti donne e uomini e le percentuali delle donne e uomini che rimangono o se ne vanno sono quasi le stesse

```python
ax = sns.countplot(x = telco_data.gender)
plt.title('Distribution of  gender')
plt.ylabel('N. Customers')
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/len(telco_data))
    x = p.get_x() + p.get_width() / 2.8
    y = p.get_y() + p.get_height() / 2
    ax.annotate(str(percentage), (x, y), color = 'white', fontsize = 12, weight = 'bold')
plt.show()
```

```python
g = telco_data.groupby('gender')['Churn']
pd.concat([g.value_counts(), g.value_counts(normalize = True).mul(100).round(1).astype(str) + '%'], axis = 1, keys = ('Count', 'Percentage'))
```

## Senior

##### Solo il 16% dei clienti sono anziani, quindi la maggior parte dei clienti nei dati sono giovani

```python
ax = sns.countplot(x = telco_data.SeniorCitizen)
plt.title('SeniorCitizen')
plt.ylabel('N. Customers')
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/len(telco_data))
    x = p.get_x() + p.get_width() / 2.7
    y = p.get_y() + p.get_height() / 2
    ax.annotate(str(percentage), (x, y), color = 'white', fontsize = 12, weight = 'bold')
plt.show()
```

##### Circa il 76% dei giovani rimane, mentre nel caso degli anziani la percentuale dei clienti che se ne va è di poco superiore a quella dei clienti che rimangono

```python
senior = telco_data.loc[(telco_data.SeniorCitizen == 1)]['Churn']
young = telco_data.loc[(telco_data.SeniorCitizen == 0)]['Churn']

fig, axes = plt.subplots(ncols = 2, figsize = (10,5))

ax1 = sns.countplot(x = senior, ax = axes[0])
ax2 = sns.countplot(x = young, ax = axes[1])
ax1.set_title('Senior')
ax1.set_ylabel('N. Senior Customers')
ax2.set_title('Young')
ax2.set_ylabel('N. Young Customers')

for p in ax1.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/len(senior))
    x = p.get_x() + p.get_width() / 2.8
    y = p.get_y() + p.get_height() / 2
    ax1.annotate(str(percentage), (x, y), color = 'white', fontsize = 12, weight = 'bold')

for p in ax2.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/len(young))
    x = p.get_x() + p.get_width() / 2.8
    y = p.get_y() + p.get_height() / 2
    ax2.annotate(str(percentage), (x, y), color = 'white', fontsize = 12, weight = 'bold')
                                   

plt.show()
```

```python
telco_data['Churn'].replace(to_replace = 'Yes', value = 1, inplace = True)
telco_data['Churn'].replace(to_replace = 'No', value = 0, inplace = True)
```

```python
telco_data.head()
```

```python
data_dummies = pd.get_dummies(telco_data)
```

```python
data_dummies.head()
```

##### Dal grafico in seguito è possibile osservare che i contratti mensili e l'assenza di servizi sono correlati positivamente al tasso di abbandono, mentre i contratti a due anni e i servizi senza connessione ad internet sono correlati negativamente con l'abbandono

```python
plt.figure(figsize = (20, 10))
data_dummies.corr()['Churn'].sort_values(ascending = False).plot(kind = 'bar')
```

#### Divisione set di dati e creazione file test.csv e train.csv

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(data_dummies, test_size=0.2)
```

```python
test.to_csv('test.csv', index=False)
train.to_csv('train.csv', index=False)
```
