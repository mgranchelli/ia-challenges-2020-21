```python
from sklearn.preprocessing import LabelEncoder

label_train_data = train_data.copy()
label_test_data = test_data.copy()

label_encoder = LabelEncoder()
label_train_data['Geography'] = label_encoder.fit_transform(train_data['Geography'])
label_train_data['Gender'] = label_encoder.fit_transform(train_data['Gender'])

label_test_data['Geography'] = label_encoder.fit_transform(test_data['Geography'])
label_test_data['Gender'] = label_encoder.fit_transform(test_data['Gender'])
```

```python
train_data.head()
```

```python
label_train_data.head()
```

```python
test_data.head()
```

```python
label_test_data.head()
```

```python
train_X = label_train_data.drop(['Exited'], axis=1)
train_y = label_train_data.Exited
```

```python
train_X.head()
```

```python
train_y.head()
```

```python
test_X = label_test_data.drop(['Exited'], axis=1)
test_y = label_test_data.Exited
```

### Train

#### LogisticRegression, è possibile osservare che si ottiene il 78% di accuracy e dal grafico è possibile osservare che la probabilità di abbandono diminuisce con il CreditScore, mentre l'età è la più alta che porta al cambio di banca 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model = LogisticRegression()
model.fit(train_X, train_y)
```

```python
prediction_test = model.predict(test_X)
print("Accuracy: " + str(accuracy_score(test_y, prediction_test)))
```

```python
weights = pd.Series(model.coef_[0],
                 index=test_X.columns.values)
print (weights.sort_values(ascending = False)[:10].plot(kind='bar'))
```

#### RandomForest, accuracy superiore a quella della LogisticRegression, pari all'86% e in questo caso in testa alle variabili predittive più importanti per prevedere il cambio banca troviamo sempre l'età, seguita dalle entrate stimate del cliente, dallo score creditizio, dal saldo e dal numero di prodotti. Come visto prima tutti i clienti con un numero di prodotti uguale a 4 cambia banca

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_rf = RandomForestClassifier()
```

```python
model_rf.fit(train_X, train_y)
```

```python
pred_y = model_rf.predict(test_X)
```

```python
print("Accuracy: " + str(accuracy_score(test_y, pred_y)))
```

```python
importances = model_rf.feature_importances_
weights = pd.Series(importances,
                 index=test_X.columns.values)
weights.sort_values()[-10:].plot(kind = 'barh')
plt.show()
```

#### Scalando tutte le variabili in un intervallo 0-1 per la LogisticRegression l'accuracy arriva all'80%, mentre per nel caso di RandomForest rimane la stessa

```python
from sklearn.preprocessing import MinMaxScaler
train_X_n = train_X.copy()
features = train_X_n.columns.values
scaler = MinMaxScaler(feature_range = (0,1))
scaler.fit(train_X_n)
train_X_n = pd.DataFrame(scaler.transform(train_X_n))
train_X_n.columns = features


test_X_n = test_X.copy()
features = test_X_n.columns.values
scaler = MinMaxScaler(feature_range = (0,1))
scaler.fit(test_X_n)
test_X_n = pd.DataFrame(scaler.transform(test_X_n))
test_X_n.columns = features
```
